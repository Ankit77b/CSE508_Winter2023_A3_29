**Creating the graph**

```python
# importing the necessary libraries..
import networkx as nx
from networkx.drawing.layout import spring_layout
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
     
```

```python
# creating the graph..
path = '/content/p2p-Gnutella04.txt'
graph = nx.read_edgelist(path,comments='#',delimiter='\t',create_using=nx.DiGraph)
# Getting the number of nodes and edges in the graph..
print(len(graph.nodes()),len(graph.edges()))
```

**Printing Page Rank Values**

```python
# getting the page rank scores for each node..
pgrank_scores = nx.pagerank(graph)
print(len(pgrank_scores))
# ranking the nodes based on their page rank scores..
dict(sorted(pgrank_scores.items(), key=lambda item: item[1], reverse = True))
```

**Distribution of Page rank scores**

```python
import matplotlib.pyplot as plt

# Get PageRank scores for each node
pgrank_scores = nx.pagerank(graph)

# Create a list of PageRank scores
scores_list = list(pgrank_scores.values())

# Plot histogram of PageRank scores
plt.hist(scores_list, bins=20, color='blue', edgecolor='black')
plt.xlabel('PageRank score', fontsize=12)
plt.ylabel('Number of nodes', fontsize=12)
plt.title('Distribution of PageRank scores', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

**Calculating hub scores and authority scores**

```python
# getting the Authority scores and Hub scores for each node 
# HITS algorithm is used here..
hub_scores, authority_scores = nx.hits(graph, max_iter = 100, normalized = True)
```

```python
# ranking the nodes based on their Authority scores..
print(len(authority_scores))
dict(sorted(authority_scores.items(), key=lambda item: item[1], reverse = True))
     
```

```python
hub_scores_dict = {k: v for k, v in sorted(hub_scores.items(), key=lambda item: item[1], reverse=True)}
top_50_hub_scores = dict(list(hub_scores_dict.items())[:50])
for node, score in top_50_hub_scores.items():
    print(f"Node {node}: Hub score = {score}")
```

```python
print(len(hub_scores))
hub_scores_dict = {k: format(v, '.20f') for k, v in sorted(hub_scores.items(), key=lambda item: item[1], reverse=True)}
print(hub_scores_dict)
```

**Plotting the Authority scores**

```python
# convert the authority scores to a list
authority_scores_list = [authority_scores[n] for n in sorted(graph.nodes)]

# create a histogram of the authority scores
plt.hist(authority_scores_list, bins=20, color='skyblue', edgecolor='black')
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Authority Scores', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Authority Scores', fontsize=14)
plt.gca().set_facecolor('#f2f2f2')
plt.show()
```

**Distribution of Hub scores**

```python
# convert the hub scores to a list
hub_scores_list = [hub_scores[n] for n in sorted(graph.nodes)]

# create a histogram of the hub scores
plt.hist(hub_scores_list, bins=20, color='skyblue', edgecolor='black')
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Hub Scores', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Hub Scores', fontsize=14)
plt.gca().set_facecolor('#f2f2f2')
plt.show()
```

**Comparison of Node Rankings for all the three scores**

```python
import matplotlib.pyplot as plt

# Get the sorted lists of node IDs for each score
pgrank_nodes = [node for node, score in sorted(pgrank_scores.items(), key=lambda item: item[1], reverse=True)]
hub_nodes = [node for node, score in sorted(hub_scores.items(), key=lambda item: item[1], reverse=True)]
authority_nodes = [node for node, score in sorted(authority_scores.items(), key=lambda item: item[1], reverse=True)]

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(range(len(pgrank_nodes)), [hub_nodes.index(node) for node in pgrank_nodes], s=20, alpha=0.5, label='HITS Hub Scores')
plt.scatter(range(len(pgrank_nodes)), [authority_nodes.index(node) for node in pgrank_nodes], s=20, alpha=0.5, label='HITS Authority Scores')
plt.scatter(range(len(pgrank_nodes)), range(len(pgrank_nodes)), s=5, alpha=0.5, label='PageRank Scores')
plt.legend()
plt.xlabel('Nodes')
plt.ylabel('Ranking')
plt.title('Comparison of Node Rankings')
plt.show()
```

```python
# convert page rank scores into dictionary
pgRank_dict = dict(enumerate(pgrank_scores))

# convert hub scores into dictionary
hub_dict = dict(enumerate(hub_scores))

# convert authority scores into dictionary
auth_dict = dict(enumerate(authority_scores))
```

```python
# sort the dictionaries in descending order of scores
pgRank_sorted = dict(sorted(pgRank_dict.items(), key=lambda x: x[1], reverse=True))
hub_sorted = dict(sorted(hub_dict.items(), key=lambda x: x[1], reverse=True))
auth_sorted = dict(sorted(auth_dict.items(), key=lambda x: x[1], reverse=True))

# get the top node for each method
pgRank_maxKey = list(pgRank_sorted.keys())[0]
hub_maxKey = list(hub_sorted.keys())[0]
auth_maxKey = list(auth_sorted.keys())[0]
```

**plotting plots for comparing the highest , average and lowest scores for all the three scores.**

```python
import numpy as np

# get highest ranked nodes for each method
pgRank_maxKey = max(pgrank_scores, key=pgrank_scores.get)
auth_maxKey = max(authority_scores, key=authority_scores.get)
hub_maxKey = max(hub_scores, key=hub_scores.get)

# get highest scores for each method
pgRank_score = pgrank_scores[pgRank_maxKey]
auth_score = authority_scores[auth_maxKey]
hub_score = hub_scores[hub_maxKey]

# get smallest scored nodes for each method
pgRank_minKey = min(pgrank_scores, key=pgrank_scores.get)
auth_minKey = min(authority_scores, key=authority_scores.get)
hub_minKey = min(hub_scores, key=hub_scores.get)

# get smallest scores for each method
pgRank_minScore = pgrank_scores[pgRank_minKey]
auth_minScore = authority_scores[auth_minKey]
hub_minScore = hub_scores[hub_minKey]

# get average scores for each method
pgRank_avgScore = np.mean(list(pgrank_scores.values()))
auth_avgScore = np.mean(list(authority_scores.values()))
hub_avgScore = np.mean(list(hub_scores.values()))

# plot for highest scored nodes
plt.figure(figsize=(5, 4))
plt.plot([int(pgRank_maxKey), int(auth_maxKey), int(hub_maxKey)], [pgRank_score, auth_score, hub_score], marker='o')
plt.text(int(pgRank_maxKey), pgRank_score, '   Page Rank Score', color="red")
plt.text(int(auth_maxKey), auth_score, '   Authority Score', color="red")
plt.text(int(hub_maxKey), hub_score, '  Hub Score', color="red")
plt.xlabel("Nodes")
plt.ylabel("Scores")
plt.title("Comparison of Scoring Methods (based on highest scores)")
plt.grid(True)
plt.show()

# plot for smallest scored nodes
plt.figure(figsize=(5, 4))
plt.plot([int(pgRank_minKey), int(auth_minKey), int(hub_minKey)], [pgRank_minScore, auth_minScore, hub_minScore], marker='o')
plt.text(int(pgRank_minKey), pgRank_minScore, '   Page Rank Score', color="red")
plt.text(int(auth_minKey), auth_minScore, '   Authority Score', color="red")
plt.text(int(hub_minKey), hub_minScore, '  Hub Score', color="red")
plt.xlabel("Nodes")
plt.ylabel("Scores")
plt.title("Comparison of Scoring Methods (based on smallest scores)")
plt.grid(True)
plt.show()

# plot for average scores
plt.figure(figsize=(5, 4))
plt.plot(["PageRank", "Authority", "Hub"], [pgRank_avgScore, auth_avgScore, hub_avgScore], marker='o')
plt.text("PageRank", pgRank_avgScore, '   Page Rank Score', color="red")
plt.text("Authority", auth_avgScore, '   Authority Score', color="red")
plt.text("Hub", hub_avgScore, '  Hub Score', color="red")
plt.xlabel("Scoring Method")
plt.ylabel("Average Score")
plt.title("Comparison of Scoring Methods (based on average scores)")
plt.grid(True)
plt.show()
```
